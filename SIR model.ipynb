## Susceptible-Infected-Recovered (SIR) model

SIR model is an epidemiological model of disease spreading

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#Initial number of infected and recovered individuals
S0 = 85.0
I0 = 10.0
R0 = 5.0

#Parameters of SIR model(per day)
b = 0.025   #conatct rate 
rho = 0.03    #recovery rate,ρ
d = 0.1     #death rate of healthy individuals
sigma = 0.3     #recovery rate of individuals susceptible of being infected again, σ
theta = 0.15    #immigration of individuals susceptible to be infected, θ
delta = 0.25    #death rate of infected individuals, δ
y0 = [85.0,10.0,5.0]

#Time array in days
t = np.linspace(0,10,101)

#defining SIR functions
def sir(y,t):
    S,I,R = y
    dSdt = theta - S - (b * I) + (sigma * R)
    dIdt = (b * S * I) - delta - (rho * I)
    dRdt = (rho * I) - (sigma * R) - (d * R)
    dydt = [dSdt, dIdt, dRdt]
    return dydt
sol = odeint(sir,y0,t)
plt.plot(t,sol[:,0],'g', label='Susceptible')
plt.text(0.2, 75, 'Susceptible')
plt.plot(t,sol[:,1],'r', label='Infected')
plt.text(7, 68, 'Infected')
plt.plot(t,sol[:,2],'b', label='Recovered')
plt.text(4, 7, 'Recovered')
plt.xlabel('Days', fontsize=12)
plt.ylabel('Number of Individuals', fontsize=12)
plt.title('SIR Model', fontsize=14)
plt.show()
```
